```python
users = interactons['user'].astype(int)
memes = interactons['meme'].astype(int)
scores = interactons['score'].astype(float)
```

```python
import scipy
import scipy.sparse.linalg

from scipy.sparse import coo_matrix
```

```python
user_meme_matrix = coo_matrix(
    (
        scores,
        (memes, users)
    )
)
```

```python
memes_factors, sigmas, users_factors = scipy.sparse.linalg.svds(user_meme_matrix, 100)
```

```python
memes_factors
```

```python
sigmas
```

```python
index = 302
meme_id, user_id, actual = memes[index], users[index], scores[index]
meme_id, user_id, actual
```

```python
(memes_factors[meme_id] * sigmas) @ users_factors.T[user_id]
```
